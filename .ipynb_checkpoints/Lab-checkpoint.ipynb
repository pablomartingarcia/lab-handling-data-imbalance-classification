```python
#1. Import the required libraries and modules that you would need.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
#2. Read that data into Python and call the dataframe churnData.

churnData = pd.read_csv('/Users/suso/BootcampDA/Unit7/lab-handling-data-imbalance-classification/files_for_lab/Customer-Churn.csv')
```

```python
#3. Check the datatypes of all the columns in the data. You would see that the column TotalCharges is object type. Convert this column into numeric type using pd.to_numeric function.

print(churnData.dtypes)
```

```python
churnData['TotalCharges'] = pd.to_numeric(churnData['TotalCharges'], errors='coerce')
```

```python
#4. Check for null values in the dataframe. Replace the null values.

print(churnData.isnull().sum())
```

```python
churnData['TotalCharges'].fillna(churnData['TotalCharges'].mean(), inplace=True)
```

```python
#5. Use the following features: tenure, SeniorCitizen, MonthlyCharges and TotalCharges:
selected_features = ['tenure', 'SeniorCitizen', 'MonthlyCharges', 'TotalCharges']
data_selected = churnData[selected_features]
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
data_scaled = scaler.fit_transform(data_selected)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data_scaled, churnData['Churn'], test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
accuracy = model.score(X_test, y_test)
print(accuracy)
```

```python
#6. Managing imbalance in the dataset

from imblearn.over_sampling import RandomOverSampler
from imblearn.under_sampling import RandomUnderSampler


oversampler = RandomOverSampler(sampling_strategy='minority')
X_train_upsampled, y_train_upsampled = oversampler.fit_resample(X_train, y_train)

undersampler = RandomUnderSampler(sampling_strategy='majority')
X_train_downsampled, y_train_downsampled = undersampler.fit_resample(X_train, y_train)

```
